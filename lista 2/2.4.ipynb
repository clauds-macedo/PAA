```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
size = np.array([i for i in range(1000, 11000, 1000)])
count = np.array([11.966, 24.303, 39.992, 53.01, 67.272, 78.692, 91.274, 113.063, 129.799, 140.538])
d = {'count':count, 'size':size}
```

```python
df = pd.DataFrame(d)
sns.set(style='darkgrid')
sns.lineplot(data=df, x=df['count'], y=df['size']).set(xlabel="count", ylabel="size")
```

## Como é possível observar no gráfico, a provável complexidade para tal algoritmo é linear, já que a quantidade de operações cresce de maneira linear em conjunto das entradas.
