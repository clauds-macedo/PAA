```python
import time
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def recursiveFib(n):
    if n <= 1:
        return n
    return recursiveFib(n-1) + recursiveFib(n-2)
recSeconds = []
for i in range (0, 40):
    recIni = time.time()
    recursiveFib(i)
    recEnd = time.time()
    recSeconds.append(recEnd - recIni)
print(recSeconds)
```

```python
fibArray = []
itrSeconds = []
for i in range(0, 40):
    itrIni = time.time()
    if i <= 1:
        fibArray.append(i)
    else:
        fibArray.append(fibArray[i-2] + fibArray[i-1])
    itrEnd = time.time()
    itrSeconds.append(itrEnd - itrIni)
```

```python
itr = np.array(itrSeconds)
rec = np.array(recSeconds)
d = {'recursive': rec, 'iterative': itr}
df = pd.DataFrame(d)
sns.set(style='darkgrid')
sns.lineplot(data=df).set(xlabel="input", ylabel="seconds")
```
